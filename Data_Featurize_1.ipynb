```python
from matminer.datasets import get_all_dataset_info

print(get_all_dataset_info("castelli_perovskites"))
```

```python
df = load_dataset("castelli_perovskites")
```

```python
df.head()
```

```python
from matminer.featurizers.composition.element import ElementFraction

ef = ElementFraction()

element_fraction_labels = ef.feature_labels()
print(element_fraction_labels)
```

```python
from matminer.featurizers.conversions import StrToComposition

stc = StrToComposition()
df1 = stc.featurize_dataframe(df, "formula", pbar=False)
```

```python
df2 = ef.featurize_dataframe(df1, "composition")

df2.head()
```

```python
from matminer.featurizers.structure import DensityFeatures

densityf = DensityFeatures()
densityf.feature_labels()
```

```python
df3 = densityf.featurize_dataframe(df2, "structure")

df3.head()
```

```python
from matminer.featurizers.composition import ElementProperty

# these are all the Magpie features. You can choose fewer if you want.
features = ['Number', 'MendeleevNumber', 'AtomicWeight', 'MeltingT', 
            'Column', 'Row', 'CovalentRadius', 'Electronegativity', 
            'NsValence', 'NpValence', 'NdValence', 'NfValence', 'NValence', 
            'NsUnfilled', 'NpUnfilled', 'NdUnfilled', 'NfUnfilled', 'NUnfilled', 
            'GSvolume_pa', 'GSbandgap', 'GSmagmom', 'SpaceGroupNumber']

# these are the statistical quantities we can compute
stats = ['mean','minimum', 'maximum', 'range', 'avg_dev', 'mode']   # we can also add 'minimum', 'maximum', 'range', 'avg_dev', 'mode'

featurizer = ElementProperty(data_source='magpie',
                             features=features,
                             stats=stats)
# featurizer = ElementProperty.from_preset('magpie')   # this is what we had last time
```

```python
# generate the features
df4 = featurizer.featurize_dataframe(df3, col_id='composition')
feature_labels = featurizer.feature_labels()   # this extracts the ugly names; but same order as "features" up above

# see the changes
df4.head()
```

```python
# generate the features
df5 = stoim.featurize_dataframe(df4, col_id='composition' )

# see the changes
df5.head()
```
